# **Compression Test of Polymer with Graphene 4% vs 9%**
#A compression test is a material characteristic test used to evaluate the behavior of a material when subjected to a compressive load. Essentially, the test measures how a material responds to forces that try to squeeze or compact it.The results from a compression test provide valuable information for engineers to determine whether a material is suitable for specific applications where compressive forces are involved, such as in structural components (beams, columns, etc.), packaging, or mechanical parts.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Compression Test GNP 4.csv')
df.head()
```

# **Strain - Stress Curve Graphene 4%**

```python
plt.figure(figsize = (12,6))
plt.plot(df.Strain, df.Stress, color='g')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 4%")
plt.show()
```

```python
df.describe()
```

# Maximum Compresive Stress(MPa) until this material fails (crack) is 0.00479 MPa

# **Strain - Stress Curve Graphene 9%**

```python
df2= pd.read_csv('Compression Test GNP 9.csv')
df2.head()
```

```python
plt.figure(figsize = (12,6))
plt.plot(df2.Strain, df2.Stress , color='b')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 9%")
plt.show()
```

```python
df2.describe()
```

# Maximum Compresive Stress(MPa) until this material fails (crack) is 0.005305 MPa

# **Graphene 4% vs 9%**

```python
plt.figure(figsize = (12,6))
plt.plot(df.Strain, df.Stress, color='g')
plt.plot(df2.Strain, df2.Stress, color='b')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 9%")
plt.show()
```

# **Summary**
# Polymer with Graphene 9% possess higher Compressive Stress compared to polymer with more Graphene (in this case 9%). We can see both from the graphs and the maximum calculated value of Stress.

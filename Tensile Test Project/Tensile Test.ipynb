# **Tensile Test of Polymer with Graphene 4% vs 9%**
# A tensile test is a fundamental material characterization technique used to assess the mechanical properties of a material when it is subjected to tension (stretching). The machine begins to pull the specimen at a constant rate, increasing the tension. As the specimen is stretched, it undergoes elastic deformation initially (proportional to the applied stress). The test continues until the material fails, meaning it fractures or breaks.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Tensile Test Data GNP 4.csv')
df.head()
```

# **Strain - Stress Curve Graphene 4%**

```python
#Data Type
type(df)
```

```python
#Plot
plt.figure(figsize = (12,6))
plt.plot(df.Strain, df.Stress, color='g')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 4%")
plt.show()
```

```python
df.describe()
```

# Maximum Tensile Stress(MPa) until this material fails (break) is 0.654 MPa

# **Strain - Stress Curve Graphene 9%**

```python
df2= pd.read_csv('Tensile Test Data GNP 9_1.csv')
df2.head()
```

```python
#Plot
plt.figure(figsize = (12,6))
plt.plot(df2.Strain, df2.Stress, color='r')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 9%")
plt.show()
```

```python
df2.describe()
```

# Maximum Tensile Stress(MPa) until this material fails (break) is 0.5148MPa

# **Graphene 4% vs 9%**

```python
plt.figure(figsize = (12,6))
plt.plot(df.Strain, df.Stress, color='g')
plt.plot(df2.Strain, df2.Stress, color='r')
plt.xlabel("Strain %")
plt.ylabel("Stress MPa")
plt.title("Strain - Stress Curve of Graphene 9%")
plt.show()
```

# **Summary**
# Polymer with Graphene 4% possess higher Tensile Stress compared to polymer with more Graphene (in this case 9%). We can see both from the graphs and the maximum value of Stress.
